```python
##Utils function
#split trainX and trainY for AR
def build_data(array,p):
  n = len(array)
  X = np.zeros((n-p,p))
  for i in range(p):
    X[:,i] = array[i:n-p+i]
  Y = array[p:]
  return X,Y

#function for regression
def regression(X,Y):
    beta = np.linalg.inv(X.T.dot(X)).dot(X.T.dot(Y))
    return beta
```

```python
##Prediction using AR and EWMA
#compute beta for AR
def Auto_Regression(array,p):
  X,Y = build_data(array,p)
  beta = regression(X,Y)
  return beta

#predict for AR
#just get last 7 days data of prediction and True value
# def AR_prediction(array,p,beta):
#   X,Y = build_data(array,p)
#   Yh = X.dot(beta)
#   return Yh[-7:],Y[-7:]

def AR_prediction(array,p,beta):
  X,Y = build_data(array,p)
  for i in range(7):
    
  return Yh[-7:],Y[-7:]

#input the whole 4 weeks data and split inpute for EWMA
#Ylen represent length of output
def EWMA(array,alpha,Ylen = 7):
  t = len(array)
  Y = array[-1*Ylen:]
  Yh = np.zeros(Ylen)
  for i in range(Ylen):
    X = array[:t-Ylen+i]
    for j in range(len(X)):
      Yh[i] += alpha*((1-alpha)**j)*X[len(X)-j-1]
  return Yh,Y
```

```python
##Function to compute MAPE and MSE
#Yh and Yt represent prediction and true value respectively

def MAPE(Yh,Yt):
  error = Yt - Yh
  mape = 0.
  length = 0.
  for i in range(len(error)):
    if Yt[i] != 0:
      mape += 100*np.abs(error[i])/Yt[i]
      length += 1
  return mape/length

def MSE(Yh,Yt):
  error = Yt - Yh
  sse = error.T.dot(error)+0.
  mse = sse/len(Yt)
  return mse
```
