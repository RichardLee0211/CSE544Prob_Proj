```python
import csv
import numpy as np
import pandas as pd
```

```python
##Load datasets
date = []
c_confirmed_CA = []
c_confirmed_CO = []
c_deaths_CA = []
c_deaths_CO = []
with open('3.csv', 'r') as read_file:
    reader = csv.reader(read_file)
    header = next(reader)
    for row in reader:
      date.append(row[0])
      c_confirmed_CA.append(int(row[1]))
      c_confirmed_CO.append(int(row[2]))
      c_deaths_CA.append(int(row[3]))
      c_deaths_CO.append(int(row[4]))

#compute the daily stats
#cdata means list of cumulative data
def compute_stats(cdata):
  daily = [cdata[0]]
  for i in range(1,len(cdata)):
    daily.append(cdata[i]-cdata[i-1])
  return daily

confirmed_CA = compute_stats(c_confirmed_CA)
confirmed_CO = compute_stats(c_confirmed_CO)
deaths_CA = compute_stats(c_deaths_CA)
deaths_CO = compute_stats(c_deaths_CO)

#report the negative data and corresponding date
#data means each list of daily stats
def report_negative(data):
  negative_list = []
  ndate = []##corresponding date
  for i in range(len(data)):
    if data[i] < 0:
      negative_list.append(data[i])
      ndate.append(date[i])
  if len(ndate) == 0:
    print('None negative value of stats!')
    return False
  else:
    print('negative value and corresponding date:')
    for i in range(len(ndate)):
      print(ndate[i])
      print(negative_list[i])
    return True
  
print('If there is negative value in CA confirmed dataset:')
print(report_negative(confirmed_CA))
print('If there is negative value in CO confirmed dataset:')
print(report_negative(confirmed_CO))
print('If there is negative value in CA deaths dataset:')
print(report_negative(deaths_CA))
print('If there is negative value in CO deaths dataset:')
print(report_negative(deaths_CO))
```

```python
##Outlier
##n is the number of  samples
n = len(date)
##Every pair of Q1,Q3 are the 25% and 75% quatiles respectively
##Every IOR is the inter-quatile range and Let alpha = 1.5(Accoding to Tukey's rule)
alpha = 1.5
##Every upper and lower define the range of outlier
##Every date is corresponding date for each dataset

#function to compute upper and lower bound of outlier detection
def outlier_range(data):
  #compute n and Q1,Q3 and IOR
  array = np.array(sorted(data))
  Q1 = array[int(np.ceil(n*0.25))-1]
  Q3 = array[int(np.ceil(n*0.75))-1]
  IOR = Q3 - Q1
  lower = Q1 - alpha*IOR
  upper = Q3 + alpha*IOR
  return lower,upper

#function to count outliers of lower and upper bound  and remove them
def outlier(data,lower,upper):
  cleaned = []
  c_date = []
  for i in range(len(data)):
    if (data[i] >= lower and data[i] <= upper) or data[i] == 0:
      cleaned.append(data[i])
      c_date.append(date[i])
  return n-len(c_date),cleaned,c_date
```

```python
#CA confirmed
print('CA confirmed dataset:')

#clean
lower1,upper1 = outlier_range(confirmed_CA)
print('lower bound of outlier = ',lower1)
print('upper bound of outlier = ',upper1)
count1,cleaned1,date1 = outlier(confirmed_CA,lower1,upper1)
print('number of outliers = ',count1)

#cleaned dataset
print(cleaned1)
print(date1)
```

```python
#CO confirmed
print('CO confirmed dataset:')

#clean
lower2,upper2 = outlier_range(confirmed_CO)
print('lower bound of outlier = ',lower2)
print('upper bound of outlier = ',upper2)
count2,cleaned2,date2 = outlier(confirmed_CO,lower2,upper2)
print('number of outliers = ', count2)

#cleaned dataset)
print(cleaned2)
print(date2)
```

```python
#CA deaths
print('CA deaths dataset:')
# array3 = np.array(deaths_CA)
# print('number of zeros = ',np.nonzero(array3)[0][0])##first none zero number

#clean
lower3,upper3 = outlier_range(deaths_CA)
print('lower bound of outlier = ',lower3)
print('upper bound of outlier = ',upper3)
count3,cleaned3,date3 = outlier(deaths_CA,lower3,upper3)
print('number of outliers = ', count3)

#cleaned dataset
print(cleaned3)
print(date3)
```

```python
#CO deaths
print('CO deaths dataset:')

#clean
lower4,upper4 = outlier_range(deaths_CO)
print('lower bound of outlier = ',lower4)
print('upper bound of outlier = ',upper4)
count4,cleaned4,date4 = outlier(deaths_CO,lower4,upper4)
print('number of outliers = ', count4)

#cleaned dataset
print(cleaned4)
print(date4)
```

```python
##Handle negative in CO deaths
#change negative value to 0
print('remained negative value and corresponding date')
for i in range(len(cleaned4)):
  if cleaned4[i] < 0:
    print(date4[i])
    print(cleaned4[i])
    cleaned4[i] = 0##replace by 0

for i in range(len(deaths_CO)):
  if deaths_CO[i] < 0:
    deaths_CO[i] = 0##replace by 0
```

```python
##Write csv data for each dataset and corresponding date
#CA confirmed
dataframe = pd.DataFrame({header[0]:date1,header[1]:cleaned1})
dataframe.to_csv("CA_confirmed_cleaned.csv",index=False)
dataframe = pd.DataFrame({header[0]:date,header[1]:confirmed_CA})
dataframe.to_csv("CA_confirmed_origin.csv",index=False)
#CO confirmed
dataframe = pd.DataFrame({header[0]:date2,header[2]:cleaned2})
dataframe.to_csv("CO_confirmed_cleaned.csv",index=False)
dataframe = pd.DataFrame({header[0]:date,header[2]:confirmed_CO})
dataframe.to_csv("CO_confirmed_origin.csv",index=False)
#CA deaths
dataframe = pd.DataFrame({header[0]:date3,header[3]:cleaned3})
dataframe.to_csv("CA_deaths_cleaned.csv",index=False)
dataframe = pd.DataFrame({header[0]:date,header[3]:deaths_CA})
dataframe.to_csv("CA_deaths_origin.csv",index=False)
#CO deaths
dataframe = pd.DataFrame({header[0]:date4,header[4]:cleaned4})
dataframe.to_csv("CO_deaths_cleaned.csv",index=False)
dataframe = pd.DataFrame({header[0]:date,header[4]:deaths_CO})
dataframe.to_csv("CO_deaths_origin.csv",index=False)
```
